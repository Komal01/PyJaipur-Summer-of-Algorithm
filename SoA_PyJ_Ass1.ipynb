**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
df['sex'].value_counts()
```

**2. What is the average age (*age* feature) of women?**

```python
women_age = df['age'][df['sex'] == 'Female']
women_age.mean()
```

```python
df.loc[df['sex'] == 'Female', 'age'].mean()
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
total_count = df['native-country'].shape[0]
Germany_count = (df['native-country'] == 'Germany').sum()
German_percentage = (Germany_count/total_count)*100
German_percentage
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
more = df.loc[df['salary']=='>50K', 'age']
less = df.loc[df['salary']=='<=50K', 'age']

print("more salary: mean {0} & std {1} | less salary: mean {2} & std {3}".format(more.mean(), more.std(), less.mean(), less.std()))
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
for (race), df_sub in df.groupby(['race']):
    for (sex), df_sub in df.groupby(['sex']):
        print('race: {0} | sex: {1}'.format(race, sex))
        print(df_sub['age'].describe())
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
df.loc[(df['sex']=='Male') & (df['marital-status'].str.startswith("Married")), 'salary'].value_counts()

df.loc[(df['sex'] == 'Male') & (df['marital-status'].isin(['Never-married', 'Separated',
                                                                'Divorced', 'Widowed'])), 'salary'].value_counts()
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
max_hrs = df['hours-per-week'].max()
print('max num hrs: {0}'.format(max_hrs))

workaholics = df[df['hours-per-week'] == max_hrs].shape[0]
print('total workers: {0}'.format(workaholics))

rich = float(df[(df['hours-per-week'] == max_hrs) & (df['salary'] == '>50K')].shape[0]) / workaholics
print('percentage: {0}'.format(int(100*rich)))
```

**10. Count the average time of work (*hours-per-week*) for those who earn a little and a lot (*salary*) for each country (*native-country*). What will these be for Japan?**

```python
for (country, salary), df_sub in df.groupby(['native-country', 'salary']):
    print(country, salary, round(df_sub['hours-per-week'].mean(),1))
```
